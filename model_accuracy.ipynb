```python
%load_ext autoreload
%autoreload 2

import numpy as np
```

```python
t50 = np.load("data/predict_feature_topo_compo_cv_50.npz")
t500 = np.load("data/predict_feature_topo_compo_cv_500.npz")
t5000 = np.load("data/predict_feature_topo_compo_cv_5000.npz")

t500_a = np.load("data/predict_feature_alpha_compo_cv_500_100.npz")
t5000_a = np.load("data/predict_feature_alpha_compo_cv_5000_100.npz")
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=[500, 5000], y=[t500["r2"], t5000["r2"]], mode='lines+markers', name='topo_compo'))
fig.add_trace(go.Scatter(x=[500, 5000], y=[t500_a["r2"], t5000_a["r2"]], mode='lines+markers', name='alpha_compo'))
fig.update_layout(title='R2 Score vs n_estimators', xaxis_title='n_estimators', yaxis_title='R2 Score')
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=[500, 5000], y=[t500["rmse"], t5000["rmse"]], mode='lines+markers', name='topo_compo'))
fig.add_trace(go.Scatter(x=[500, 5000], y=[t500_a["rmse"], t5000_a["rmse"]], mode='lines+markers', name='alpha_compo'))
fig.update_layout(title='RMSE vs n_estimators', xaxis_title='n_estimators', yaxis_title='RMSE')
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=[500, 5000], y=[t500["time"] / 60, t5000["time"] / 60], mode='lines+markers', name='topo_compo'))
fig.add_trace(go.Scatter(x=[500, 5000], y=[t500_a["time"] / 60, t5000_a["time"] / 60], mode='lines+markers', name='alpha_compo'))
fig.update_layout(title='Time vs n_estimators', xaxis_title='n_estimators', yaxis_title='Time (minutes)')
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=t5000["y"], y=t5000["predicted"], mode='markers', name='topo_compo'))
fig.add_trace(go.Scatter(x=t5000_a["y"], y=t5000_a["predicted"], mode='markers', name='alpha_compo'))
fig.add_trace(go.Scatter(x=[min(t5000["y"].min(), t5000_a["y"].min()), max(t5000["y"].max(), t5000_a["y"].max())], y=[min(t5000["y"].min(), t5000_a["y"].min()), max(t5000["y"].max(), t5000_a["y"].max())], mode='lines', name='Ideal Fit', line=dict(dash='dash')))
fig.update_layout(title='Actual vs Predicted', xaxis_title='Actual', yaxis_title='Predicted')
```
