```python
import pandas as pd
import numpy as np
import kmapper as km
df = pd.read_csv("mapper_data.csv")
ids = df['id']
df = df.drop('id', axis=1)
data = df.to_numpy()
mapper = km.KeplerMapper(verbose=1)
projected_data = mapper.fit_transform(data, projection="[0,1]") 
cover = km.Cover(n_cubes=100)
graph = mapper.map(projected_data, data, cover=cover)
mapper.visualize(graph, path_html="visualization.html", custom_tooltips=ids,
                 title="data kepler mapper")
```
